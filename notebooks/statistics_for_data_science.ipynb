This notebook provides practical implementations of key **Statistics** concepts used in **Data Science**.
Topics covered:

- Descriptive Statistics

Let's explore these concepts with Python! ðŸš€


```python
# Importing essential libraries
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

## ðŸ“ˆ Descriptive Statistics

```python
# Generating synthetic data
data = np.random.normal(loc=50, scale=15, size=1000)  # Mean 50, Std Dev 15

# Computing descriptive statistics
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data, keepdims=True).mode[0]
std_dev = np.std(data)
variance = np.var(data)

# Displaying results
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Variance: {variance:.2f}")
```

## ðŸŽ² Probability Distributions

```python
# Generating a normal distribution and visualizing it
sns.histplot(data, bins=30, kde=True)
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

## ðŸ§ª Hypothesis Testing

```python
# Performing a t-test for two samples
group_a = np.random.normal(50, 10, 100)
group_b = np.random.normal(52, 10, 100)

stat, p_value = stats.ttest_ind(group_a, group_b)
print(f"T-test - Statistic: {stat:.2f}, P-value: {p_value:.4f}")
```

## ðŸ”„ Correlation & Regression

```python
# Creating a dataset with a linear relationship
np.random.seed(42)
x = np.random.rand(100, 1) * 10
y = 2.5 * x + np.random.randn(100, 1) * 2

# Creating and training a linear regression model
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

# Plotting the regression
plt.scatter(x, y, label="Actual Data")
plt.plot(x, y_pred, color='red', label="Linear Regression")
plt.xlabel("Independent Variable (x)")
plt.ylabel("Dependent Variable (y)")
plt.title("Simple Linear Regression")
plt.legend()
plt.show()
```
